# Step 5: Distribution Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# set up the plot styles
sns.set(style="whitegrid")

# histogram for Viseu using ('temperature_2m_mean') column
plt.figure(figsize=(10, 6))
sns.histplot(viseu_df['temperature_2m_mean'], kde=True, bins=30, color="blue", label="Viseu")
plt.title('Temperature Distribution for Viseu')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# histogram for Bragança using ('temperature_2m_mean') column
plt.figure(figsize=(10, 6))
sns.histplot(braganca_df['temperature_2m_mean'], kde=True, bins=30, color="red", label="Bragança")
plt.title('Temperature Distribution for Bragança')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# boxplot comparison for viseu and bragança
plt.figure(figsize=(10, 6))

# combine data into a single dataFrame for easy plotting
combined_data = pd.DataFrame({
    'Temperature': pd.concat([viseu_df['temperature_2m_mean'], braganca_df['temperature_2m_mean']]),
    'Location': ['Viseu'] * len(viseu_df) + ['Bragança'] * len(braganca_df)
})

# create the boxplot using the combined DataFrame
sns.boxplot(x='Location', y='Temperature', data=combined_data)
plt.title('Temperature Comparison: Viseu vs Bragança')
plt.ylabel('Temperature (°C)')
plt.show()
```

# Step 6: Monthly or Seasonal Aggregation

```python
# convert the 'date' column to datetime if it's not already
braganca_df['date'] = pd.to_datetime(braganca_df['date'])
viseu_df['date'] = pd.to_datetime(viseu_df['date'])

# extract month or season
braganca_df['Month'] = braganca_df['date'].dt.month
viseu_df['Month'] = viseu_df['date'].dt.month

# monthly temperature aggregation for both locations
monthly_braganca = braganca_df.groupby('Month')['temperature_2m_mean'].mean()
monthly_viseu = viseu_df.groupby('Month')['temperature_2m_mean'].mean()

# plot monthly average temperature
plt.figure(figsize=(10, 6))
plt.plot(monthly_braganca.index, monthly_braganca.values, label='Bragança', color='red')
plt.plot(monthly_viseu.index, monthly_viseu.values, label='Viseu', color='blue')
plt.title('Monthly Average Temperature: Viseu vs Bragança')
plt.xlabel('Month')
plt.ylabel('Average Temperature (°C)')
plt.legend()
plt.show()
```

# Step 7: Identify Patterns in Temperature Variation

```python
# standard deviation to measure variability using the 'temperature_2m_mean' column
braganca_std = braganca_df['temperature_2m_mean'].std()
viseu_std = viseu_df['temperature_2m_mean'].std()

print(f"Bragança Temperature Standard Deviation: {braganca_std}")
print(f"Viseu Temperature Standard Deviation: {viseu_std}")
```
